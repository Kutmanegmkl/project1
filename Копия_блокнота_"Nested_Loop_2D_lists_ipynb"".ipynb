# Вложенный цикл

image.png

````python
word_1 = ['Ай', 'Батыр', 'Бай', 'Кадыр', 'Назар', 'Манас', 'Мырза']
word_2 = ['бек', 'арыс', 'султан', 'хан', 'жан', 'жол', 'темир']
````

Даны части слов, из которых можно сконструировать казахские имена. Давайте отобразим все возможные сочетания. Для этого нам понадобится вложенный цикл.

````python
for i in word_1:
  for j in word_2:
    print(i+j)
````

Вероятно, некоторые имена не употребимы в таком виде, но мы наглядно продемонстрировали идею перебора всех возможных сочетаний элементов из двух множеств.

Другой пример создания числовой таблицы.

````python
x1 = [1,2,3,4]
x2 = [5,6,7,8]
x3 = [x1,x2]
print(*x3, sep='\n')
````

````python
for i in range(len(x3)):
  for j in range(len(x3[i])):
    print(x3[i][j])
````


# Generation

Создадим матрицу из нулей размером 5х5 классическим методом.

````python
res = [] # глобальный лист.
for i in range(5): # 5 итераций при каждой из которых мы будем создавать вложенный лист.
  temp = [] # создаем новый вложенный лист
  for j in range(5): # 5 итераций при каждой из которых мы будем добавлять ноль во вложенный лист.
    temp.append(0) # наполнили вложенный лист
  res.append(temp) # загружаем вложенный лист в глобальный
````

````python
res
````

Видим, что для вложенного списка мы использовали вложенный цикл.

Данный код можно немного упростиь используя list comprehention для генерации каждого вложенного списка.

````python
res = []
for i in range(5):
  res.append([0 for j in range(5)])
res
````

Теперь немного треша с list comprehention 2D:

````python
[[0 for i in range(5)] for j in range(5)]
````

Сгенерим числовую квадратную матрицу чисел от 1 до n.

Пример:



```

code...

```



````python
n = 9

for i in range(1, 9, 3):
  print(i)
````

Вывели значение первого столбца.

````python
n = 16
res = []

for i in range(1, n, int(n**0.5)):
  temp = []
  for j in range(i, i+int(n**0.5)):
    temp.append(j)
  res.append(temp)
````

````python
print(*res, sep='\n')
````

Напишем код короче

````python
n = 9
res = []

for i in range(1, n, int(n**.5)):
  res.append([j for j in range(i, i+int(n**.5))])
res
````

````python
n = 25
[[j for j in range(i, i+int(n**.5))] for i in range(1, n, int(n**.5))]
````

Оформим код как функцию

````python
def matrix(n):
  return [[j for j in range(i, i+int(n**.5))] for i in range(1, n, int(n**.5))]
````

````python
matrix(4)
````

````python
matrix(16)
````
