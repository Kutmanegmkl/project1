```
input:
['A', 'B' ...]

output:
['A_1', 'A_2', 'A_3', 'A_4', 'A_5', 'B_1' ...]
```



````python
n=5
uniq_list = list('ABCDE')
uniq_list
````

````python
dupl_list = [[f'{ltr}_{num}' for num in range(1, n+1)] for ltr in uniq_list]
dupl_list
````

````python
res = []
for ltr in uniq_list:
  res += [f'{ltr}_{num}' for num in range(1, n+1)]
print(res)
````

# Task_2

    .      ,             .       .



```
output:

['January, 2015', 'January, 2016', ... , 'December, 2020']
```



````python
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
years = [str(i) for i in range(2015, 2021)]
````

````python
[m+', '+str(year) for year in years for m in months]
````

````python
dates = []
for year in years:
  for month in months:
    dates.append(month + ', ' + year)

print(dates)
````
