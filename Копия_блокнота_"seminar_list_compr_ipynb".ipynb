# Task_18

Дан массив чисел. Выполните нормализацию списка, то есть приведите все числа в диапазон значений $[0; 1]$. Для этого можно воспользоваться формулой:

$$x_{norm} = \frac{x_i - x_{min}}{x_{max} - x_{min}}$$

# Task_19*

Дан массив чисел. Выполните его стандартизацию. Для этого воспользуйтесь формулами:

$$x_{std} = \frac{x_i-\bar{x}}{\sigma_x}$$

Где $\bar{x}$ - среднее арифметическое.

$\sigma_x$ - стандартное отклонение, которое можно вычислить по формуле:

$$D(X) = \frac{1}{n} \displaystyle\sum_{i=1}^{n} (x_i-\bar{x})^2$$

$$\sigma_x = \sqrt{D(X)}$$

````python
list_ = [150, 149, 178, 170, 200, 198, 189, 161, 171, 164, 195, 186]
x_avg=sum(list_)/len(list_)
summ=0
````

````python
for x in list_:
  summ=summ+(x-x_avg)**2
````

````python
smx=(summ/len(list_))**0.5
````

````python
xstd=[(x-x_avg)/smx for x in list_]
````

````python
xstd
````

````python
round((sum(xstd)/len(xstd)))
````
